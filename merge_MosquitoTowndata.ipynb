### Merge mosquito data (tested and not tested) to town shapefile

```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot
from shapely.geometry import LineString
from descartes import PolygonPatch
```

```python
gdf2 = gdf.drop_duplicates(subset ="TOWN_ID", keep = 'first',inplace=False) 
```

```python
gdf2.shape
```

```python
mosdf = pd.read_csv("/Users/lbrown01/Dropbox/DataScienceStuff/Data/WNV_EE/formatMos.csv")
```

```python
mosdf.head(5)
```

```python
#make a new column for town to match town shapefile
mosdf['TOWN'] = mosdf['Town']
```

```python
#capitalize all town names in the new column to match town shapefile
mosdf['TOWN'] = mosdf['TOWN'].str.upper()
```

```python
mosdf.head(5)
```

```python
mosdf2 = pd.merge(gdf2,mosdf,on="TOWN",how="right")
```

```python
#write to file
mosdf2.to_csv("/Users/lbrown01/Dropbox/DataScienceStuff/Data/WNV_EE/MosTowndat.csv", encoding='utf-8', index=False)
```
